```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('result-all-500.csv')
```

```python
df.describe()
```

```python

fig, ax = plt.subplots(figsize=(10, 6))

# 
ax.plot(df['faster-rcnn-pc'], label='faster-rcnn-pc', color='red')
ax.plot(df['yolov8n-pc'], label='yolov8n-pc', color='blue')

# 
ax.set_title('Performance Comparison')
ax.set_xlabel('Sample Index')
ax.set_ylabel('predict time(s)')

# 
ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

# 
ax.plot(df['yolov5'], label='yolov5', color='red')
ax.plot(df['yolov8n'], label='yolov8n', color='blue')
ax.plot(df['yolov8n-gray'], label='yolov8n-gray', color='gray')
ax.plot(df['yolov8n-sparsified'], label='yolov8n-sparsified', color='green')

# 
ax.set_title('Performance Comparison')
ax.set_xlabel('Sample Index')
ax.set_ylabel('predict time(s)')

# 
ax.legend()
```
